```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,8])
y=np.array([3,10])
plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,8])
y=np.array([3,10])
plt.plot(x,y,'o')
plt.show()
```

```python
x=np.array([1,8,4,3,9])
y=np.array([3,2,5,3,7])
plt.plot(x,y)
plt.show()
```

```python
y=np.array([3,10,3,6])
plt.plot(y)
plt.show()
```

```python
y=np.array([3,10,3,6])
plt.plot(y,marker='o')
plt.show()
```

```python
y=np.array([3,10,3,6])
plt.plot(y,'o:r')
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y,marker='o')
plt.show()
```
